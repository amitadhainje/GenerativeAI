# PromptTemplate for explaining any terms

```python
demo_template = '''I want you to act as a financial advisor for people. In an easy way and layman terms, please explain the {financial_concept} '''
prompt = PromptTemplate(
    input_variables = ['financial_concept'],
    template = demo_template
)
prompt.format(financial_concept="Income Tax")
```

```python
llm = ChatOpenAI(model="gpt-4o-mini")
chain = llm | prompt


result = chain.invoke("GST")
print(result.text)
```

# PromptTemplate for Language Translation

```python
language_template = "I want you to become a translator and translate the given '{sentence}' into the {target_language}"
language_prompt = PromptTemplate(
    input_variables = ['sentence',"target_language"],
    template = language_template
)
language_prompt.format(sentence="How are you", target_language="Marathi")
```

```python
chain = language_prompt | llm | StrOutputParser()

response = chain.invoke({
    "sentence": "Hello ... How are you ",
    "target_language": "Marathi"
})
print (response)
```

# FewShotTemplate for Sentimment Analysis

```python
examples = [
    {"text": "I love this product!", "sentiment": "Positive"},
    {"text": "This is the worst experience ever.", "sentiment": "Negative"},
    {"text": "It's okay, not great.", "sentiment": "Neutral"},
]
example_prompt = PromptTemplate(
    input_variables=["text", "sentiment"],
    template="Text: {text}\nSentiment: {sentiment}"
)

few_shot_prompt = FewShotPromptTemplate(
    examples=examples,
    example_prompt=example_prompt,
    prefix="Classify the sentiment of the following text.",
    suffix="Text: {input}\nSentiment:",
    input_variables=["input"],
)

chain = few_shot_prompt | llm | StrOutputParser()

result = chain.invoke({"input": "The service was fast and friendly"})
print("Following is the sentiment -- ",result)
```
